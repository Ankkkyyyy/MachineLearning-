```python
# from a pandas dataframe 

batsman = pd.read_csv('dataset/sharma-kohli.csv')
batsman
```

```python
# x axis pe season y pe runs
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title("Rohit Sharma vs Virat Kohli in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')
plt.title("Rohit Sharma vs Virat Kohli in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

## Parameters
1. color = 
2. linestyle = ''  options -> solid , dashed,dotted, dashdot
3. line width = 
4. marker = D,+
5. markersize = 

6. label='' & then call plt.legend()  for boxing


```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed', linewidth=3,marker='+',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot')
plt.title("Rohit Sharma vs Virat Kohli in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed', linewidth=3,marker='+',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',label='Rohit')
plt.title("Rohit Sharma vs Virat Kohli in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
```

### Limiting axes 

```python
# limiting axes  
price = [48000,54000,57000,49000,47000,450000] # Outlier hai last ka so graph ka baki ka curve will look flat 
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

#### So in above e.g their is a Outlier as last element so graph ka baki ka curve will look flat 
so to deal with that we will trim from x axis 

```python
price = [48000,54000,57000,49000,47000,450000] 
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)

plt.ylim(0,100000)
plt.xlim(2017,2019)
```

## Scatter Plot

1. It is just lyk a 2-D plot we use scatter plot for Bivariate Analysis 
2. Numerical vs Numerical Column pe hie use krte mostly 
3. Use-Case - Finding Correlation ,  ( For finding correlation between 2 quantities)

```python
# plt.scatter 

x = np.linspace(-10,10,50)
y = 10*x +3 + np.random.randint(0,300,50) # y = max + b + random number  
plt.scatter(x,y)
```

## Bar Chart/Graph   --   plt.bar()

x-axis pe categorical column ko rkhte hai & 

y-axis pe aggregate kindoff operation perform karte hai 

1. Bivariate Analysis 
2. Univariate Analysis
3. Generally its Numerical vs Categorical  
4. Use Case : Aggregate Analysis of a Groups 

```python
# Simple bar chart 

children = [10,20,100,20,10]
colors = ['red','blue','orange','green','yellow']
plt.bar(colors,children,color='red')
plt.xlabel('colors')
plt.ylabel('No. of likes by children')
```

### Horizontal Bar Chart  -- plt.barh()


```python
children = [10,20,100,20,10]
colors = ['red','blue','orange','green','yellow']
plt.barh(colors,children,color='red')
plt.xlabel('colors')
plt.ylabel('No. of likes by children')
```

```python
df = pd.read_csv('dataset/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],color='brown')
```

### Multiple Bar Chart 

```python
# plt.bar(df['batsman'],df['2015'],width=0.3)
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='purple')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

## Pie Chart 

1. Univariate & Bivariate Analysis dono pe kaam aaata hai 
2. categorical , Numerical dono  tareh ke data pe kaam aata hai 
3. Use Case - To find contribution on a standard scale (100%)

```python
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
data = [23,45,100,20,49]
subjects = ['sanskirt','french','Eng','Marathi','German']
plt.pie(data,labels =subjects)
plt.show()
```

```python
gayle = pd.read_csv('dataset/gayle-175.csv')
gayle
```

### autopct='%0.1f%%'
use for showing percentage in pie charts

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',colors=['Blue',"green",'yellow','brown','red','black'])
plt.show()
```

### Explode explode=[]
particular pie ko extract karsakte hai hum & alag dikhasakte hai 

### shadow = True 
thoda 3d type feel aata hai bss

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',explode=[0.4,0,0,0,0,0],shadow=True)
plt.show()
```
