# Scrapingg AJIO 

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
```

```python
s=Service("C:/Users/hp/Desktop/chromedriver.exe")
```

```python
driver = webdriver.Chrome(service=s)
driver.get("https://www.ajio.com/search/?text=backpack")

old_height = driver.execute_script('return document.body.scrollHeight')# first time mai end tak jitna page load hua uska height batata hai

counter = 1
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)') 
    time.sleep(2)
    new_height =  driver.execute_script('return document.body.scrollHeight')

    print("Counter : ",counter)
    counter+=1
    print("Old Height :",old_height)
    print("New Height :",new_height)
    # jaise hie data khtm hoga at the end so hum dekhenge ki old height jo hai kya vo equal hai new height ke tou vaise hie um break krdenge
    if new_height==old_height:
        break

    old_height=new_height


# print(height)

html = driver.page_source
with open('ajio.html','w',encoding='utf-8') as f:
    f.write(html)
```
