# *Scrapping Smartprix* 

```python
import pandas as pd
```

```python
with open('smartprix.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
soup.prettify()
```

```python
len(soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'}))
```

```python
container = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
name = []
price = []
specs = []
sim = []
processor = []
ram=[]
battery = []
display = []
camera = []
os = []
for i in container:
#     print(i.find('h2').text)
    name.append(i.find('h2').text)
#     print(i.find('span',{'class':'price'}).text)
    price.append(i.find('span',{'class':'price'}).text)
#     print(i.find('div',{'class':'score'}).find('b').text)
    specs.append(i.find('div',{'class':'score'}).find('b').text)
    
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    sim.append(x[0].text)
    processor.append(x[1].text)
    ram.append(x[2].text)
    battery.append(x[3].text)
    display.append(x[4].text)
    camera.append(x[-3].text)
    os.append(x[-1].text)
    
    
```

```python
len(name)
```

```python
len(price)
```

```python
len(specs)
```

```python
df=pd.DataFrame({
    'Model':name,
    'Price':price,
    'Rating':specs,
    'Sim':sim,
    'Processor':processor,
    'Ram':ram,
    'Battery':battery,
    'Display':display,
    'Camera':camera,
    'Operating System':os
})
```
