## Qs. Action Movie with rating higher than 7.5

```python
movies.head(2)
```

```python
movies['genres']
```

```python
action_movies = movies['genres'].str.split('|').apply(lambda x:'Action' in x)
```

```python
action_rating= movies['imdb_rating'] > 7.5
```

```python
movies[action_movies & action_rating]
```

alternative solution with help of contains() 

```python
action_movies2 = movies['genres'].str.contains('Action')
action_rating2 = movies['imdb_rating'] > 7.5
movies[action_movies2 & action_rating2]
```
