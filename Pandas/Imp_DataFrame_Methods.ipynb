#### apply():  Used in Series & Dataframe both  * apply() ki khaash baat yeah hai ki isme hum apna custom logic or function deskte 

Series mai apply krte waqt row by row lete hai & perform krte hai  ( basically column ko traverse karte hai )

Dataframe ke time row by row traverse krte by putting axis =1 


```python
# Series 
tempaply = pd.Series([10,9,9,99])

tempaply
```

```python
def sigmoid(value):
    return 1/1+np.exp(-value) 
```

```python
tempaply.apply(sigmoid)
```

```python
# Dataframe 

points_df = pd.DataFrame(
{
  '1st point':[(3,4),(-9,5),(0,0),(-10,1),(9,99)],
  '2nd point':[(-3,4),(9,5),(0,0),(10,10),(1,1)]
}
)

points_df
```

```python
# Euclid distance nikalenge 1st point ka 1 row leke then 2nd row ka 1 point lekr

def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    return ((pt_A[0] - pt_B[0])**2  + (pt_A[1] - pt_B[1])**2)**0.5

points_df.apply(euclidean,axis=1)
# axis =1 mtlb row bhejrahe hai hum
```

```python
points_df['Distance']= points_df.apply(euclidean,axis=1)
points_df
```
