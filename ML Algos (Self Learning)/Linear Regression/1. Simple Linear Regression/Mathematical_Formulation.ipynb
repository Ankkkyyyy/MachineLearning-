1.) Closed Formed Solution 

kisibhi chiz ka mathematic formula banalete hai agr tou vo closed form solution hota hai.. yahape differentiation,integration nahi hota hai.

a. Ordinary Least Square    -> Scikit Learn  andar yeahi use karta hai 


2.) Non-Closed Formed Solution 

approximation based technique hai jaha humme differentiation,Integration use krke krte hai 

a.Gradient Descent.


SGDRegressor ---> Gradient Descent use karta hai internally 

Jbh hum Higher Dimension mai kaam karte hai vaha pe humme OLS use karke m&p nikalna mushkil hojata hai    

tou vahape humme Gradient Descent use karna padta hai

```python
class HumaraLR:
    def __init__(self):
        self.m = None 
        self.b = None 
    
    def fit(self,x_train,y_train):
        # Model ko train krna hai yaha 
        # LR mai Model ko train krna mtlb m & b ko calculate karna 
        numerator = 0 
        denominator = 0 
        for i in range(x_train.shape[0]):

            numerator = numerator + (x_train[i]- x_train.mean()) * (y_train[i]-y_train.mean())
            denominator = denominator + (x_train[i]- x_train.mean()) *(x_train[i]- x_train.mean())
        
        self.m = numerator/denominator
        self.b = y_train.mean() - self.m * x_train.mean() 
        print(self.m)
        print(self.b)
    

    def predict(self,X_test):
        # y = mx + b
        return self.m*X_test+self.b
    
```
